```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
df = pd.read_csv('data.csv')
x = df['X']
y = df['Y']
```

```python
plt.figure(figsize=(12, 6))
plt.scatter(x, y, s=100, alpha=0.6)
plt.xlabel('X values', fontsize=12)
plt.ylabel('Y values', fontsize=12)
plt.title('Scatter plot of X vs Y', fontsize=14)
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
xMean, yMean = np.mean(x), np.mean(y)
numerator = np.sum((x[i] - xMean) * (y[i] - yMean) for i in range(len(x)))
denominator = np.sum((x[i] - xMean) ** 2 for i in range(len(x)))
m = numerator / denominator
c = yMean - m * xMean

def predict(x):
    return m * x + c
y_pred = predict(x)
plt.figure(figsize=(12, 6))
plt.scatter(x, y, s=100, alpha=0.6, label='Data points')
plt.plot(x, y_pred, color='red', linewidth=2, label='Regression line')
plt.xlabel('X values', fontsize=12) 
plt.ylabel('Y values', fontsize=12)
plt.title('Linear Regression Fit', fontsize=14)
plt.legend()
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
r_squared = 1 - (np.sum((y[i] - y_pred[i]) ** 2 for i in range(len(y))) / np.sum((y[i] - yMean) ** 2 for i in range(len(y))))
print(f'R-squared: {r_squared:.4f}')
```
